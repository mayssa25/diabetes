Cet ensemble de données provient à l'origine de l'Institut national du diabète et des troubles digestifs et rénaux

## Objectif à atteindre
Dans cette étude, nous avons fait des analyses sur l'ensemble de données de diabète pour avoir la relation entre l'age et le diabète.

## les 9 attributs:

Grossesses : Pour indiquer le nombre de grossesses

Glucose : Pour exprimer les niveaux de glucose dans le sang

BloodPressure : pour indiquer les mesures de la pression artérielle

SkinThickness : Pour exprimer l'épaisseur de la peau

IMC : Pour exprimer l'indice de masse corporelle

Résultat : Pour énoncer le résultat final 1 est Oui et 0 est Non

```python
x= df.iloc[:, 7].values
```

```python
def segment_age(x):
    segment = ''
    if x["Age"] >= 21 and x["Age"] < 40:
        segment = "Young Adults [21,40)"
    elif x["Age"] >= 40 and x["Age"] < 60:
        segment = "Middle Adults [40,60)"
    elif x["Age"] >= 60 and x["Age"] <= 81:
        segment = "Old Adults [60,81)"
    return segment

df["Age"] = df.apply(lambda x : segment_age(x), axis=1)
```

```python
df["Age"]
```

```python
diabetes_by_Glucose = pd.pivot_table(df, values='Glucose', index='Age', columns='Outcome', aggfunc=len).fillna(0.0)
diabetes_by_Glucose.columns = ['Not Diabetes', 'Diabetes']
```

```python
diabetes_by_Glucose
```

```python
diabetes_by_Glucose.sum()
```

```python
diabetes_by_Glucose.plot(
    kind='bar',
    figsize=(15, 7),
    title='Diabetes by Age',
    grid=True,
)

mtp.show()
```
